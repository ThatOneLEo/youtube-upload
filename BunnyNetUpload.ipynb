```python
# Install the requests library
!pip install requests

import os
import requests

# Mount Google Drive
# from google.colab import drive
# drive.mount('/content/drive')

# API key and library ID
api_key = "e4f45eae-58c9-49fa-83a16bd0842c-343a-4210"
library_id = "271653"  # Your actual library ID
collection_id = "9b4661f2-da32-48ab-9c98-77a4a0ef5e8c"  # Your collection ID
directory_path = "/content/drive/MyDrive/Cerebellum Pharmacology"

def create_video_entry(title, collection_id):
    create_video_url = f"https://video.bunnycdn.com/library/{library_id}/videos"
    create_video_headers = {
        "Accept": "application/json",
        "Content-Type": "application/json",
        "AccessKey": api_key
    }
    create_video_data = {
        "title": title,
        "collectionId": collection_id
    }

    response = requests.post(create_video_url, headers=create_video_headers, json=create_video_data)
    if response.status_code == 200:
        return response.json()["guid"]
    else:
        print(f"Failed to create video entry for {title}. Status code: {response.status_code}")
        print(response.text)
        return None

def upload_video(video_id, video_file_path):
    upload_video_url = f"https://video.bunnycdn.com/library/{library_id}/videos/{video_id}"
    upload_video_headers = {
        "AccessKey": api_key,
        "Content-Type": "application/octet-stream"
    }

    with open(video_file_path, 'rb') as video_file:
        upload_response = requests.put(upload_video_url, headers=upload_video_headers, data=video_file)

    if upload_response.status_code == 200:
        print(f"Video {video_file_path} uploaded successfully.")
    else:
        print(f"Failed to upload video {video_file_path}. Status code: {upload_response.status_code}")
        print(upload_response.text)

def upload_videos_in_directory(directory_path):
    for root, dirs, files in os.walk(directory_path):
        for file in files:
            if file.lower().endswith(('.mp4', '.mov', '.avi', '.mkv', '.flv')):  # Add other video formats if needed
                video_file_path = os.path.join(root, file)
                video_title = os.path.splitext(file)[0]
                print(f"Processing file: {video_file_path} with title: {video_title}")
                video_id = create_video_entry(video_title, collection_id)
                if video_id:
                    upload_video(video_id, video_file_path)

# Start the upload process
upload_videos_in_directory(directory_path)
```
